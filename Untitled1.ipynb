```python
import pandas as pd
import numpy as np
import math
from pandas import Series,DataFrame
import geopandas as gpd
from datetime import datetime, date, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('./housing_in_london.csv')
df['date'] = pd.to_datetime(df['date'])
pd.set_option('display.max_columns', None)
```

```python
'''
1. what does the market look like now?
2. what has the best growth potential?
3. what effect did brexit have?
'''
```

```python
# creating dfs for regions and london boroughs
regions = df[df['borough_flag'] == 0]
london = df[df['borough_flag'] == 1]
```

```python
# importing shapefiles
regions_gdf = gpd.read_file('./regions/88453ea6-7fcc-433e-9b72-4d628c7b4893202048-1-1ku2t1w.tzaug.shp')
london_gdf = gpd.read_file('./london-wards-2018/london-wards-2018_ESRI/London_Ward_CityMerged.shp')
```

```python
# creating merged dfs to plot shapefiles
def merge_dfs(df1,df2):
    '''
    input:  pandas dataframe and GeoDataFrame
    output: merged GeoDataFrame where geometry has been added
            to pandas dataframe
    '''
    if 'rgn17cd' in df2.columns.values:
        print('in columns')
        codes = 'rgn17cd'
    else:
        print('not in columns')
        codes = 'LAGSSCODE'
    
    recent_df = df1[df1['date'] == '2019-11-01']
    merged_df = recent_df.merge(df2[[codes,'geometry']], left_on='code', right_on=codes).drop('date', axis=1)
    merged_gdf = gpd.GeoDataFrame(merged_df)

    return merged_gdf
```

```python
regions_merged = merge_dfs(regions,regions_gdf)
london_merged = merge_dfs(london,london_gdf)
```

```python
# plotting shapefiles
def plot_maps(df,cat):
    '''
    input:  GeoDataFrame and category
            For average price, enter "price" in cat
            argument. For number of houses sold, enter "volume"
            in cat argument.
    output: spatial plot showing either price or volume in
            specified map
    '''
    # setting up figure
    fig, ax =  plt.subplots(1,1)
    ax.set_axis_off()
    
    # deciding which plot to create
    if cat == 'price':
        column = 'average_price'
        df.plot(column, ax=ax, legend=True, cmap='plasma', legend_kwds={'label':"Average Price"})
        title  = 'Average Price'
    elif cat == 'volume':
        column = 'houses_sold'
        df.plot(column, ax=ax, legend=True, cmap='plasma', legend_kwds={'label':"No of Houses Sold"})
        title = 'No of Houses Sold'
    else:
        print('unable to plot map')
        
    plt.title(title)
    plt.show
```

```python
plot_maps(regions_merged,'price')
```

```python
'''questions 2 & 3'''
# create dataframes for dates 11/09 to 11/19
london_growth_df = london[(london['date'] >= '2009-11-01') & (london['date'] <= '2019-11-01')].reset_index(drop=True)
regions_growth_df = regions[(regions['date'] >= '2009-11-01') & (regions['date'] <= '2019-11-01')].reset_index(drop=True)
```

```python
regions_growth_df['date'] = pd.to_datetime(regions_growth_df['date'])
regions_growth_df['date'] = regions_growth_df['date'].dt.date
```

```python
def plot_growth(df,area,brexit):
    '''
    input: a pandas dataframe,
           a string denoting the area that data is being plotted for,
           true or false to plot date of brexit vote, where true creates grey line
    output: A plot showing average price against the number of houses sold for
            a specific area. If True is selected, the plot will also show a grey
            line denoting the date of the brexit vote.
    '''
    # create dataframes
    df_new = df[df['area'] == area][['date','area','average_price','houses_sold']]
    df_new.set_index(['date'], inplace=True)
    
    # create plots()
    f,ax1 = plt.subplots()
    ax2 = ax1.twinx()
    ax1.plot(df_new['average_price'],color='blue',label='average price')
    ax2.bar(df_new.index,df_new['houses_sold'],width=20,alpha=0.2,color='coral',label='houses sold')

    # format plots using area in title
    ax1.set_title('Average Price vs. Houses Sold - ' + area.title())
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Average Price (in Â£)')
    ax2.set_ylabel('No of Houses Sold')
    ax1.set_xlim('2009-11-01','2019-11-01')
    
    # add legend
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper left')    
        
    # check if Brexit vote date should be marked
    if brexit == True:
        x = '2016-06-23'
        plt.axvline(x=x,ls='--',c='grey')
```

```python
plot_growth(regions_growth_df,'england',True)
```
